## Phase 1: Deep EDA — Understanding the Data

Before building ANY model, we need to deeply understand what we're working with. This phase answers critical questions:

- **How many emails** are in each dataset? What columns do they have?
- **Are there missing values?** Where and how many?
- **Is the data balanced?** Equal phishing vs. legitimate? (Critical — imbalanced data breaks naive accuracy metrics)
- **Do phishing emails differ in length** from legitimate ones? (Length itself can be a feature)

### Why this matters
Every assumption made here propagates through the entire pipeline. If you miss a null-handling issue now, your model will silently fail later. If you don't check class balance, you might train a model with "95% accuracy" that never actually catches a single phishing email.

```python
# 1.1 — Shape and column overview for every dataset
# This tells us: how large is each dataset, and do the schemas match?

datasets = {
    'Enron': df_enron,
    'Ling': df_ling,
    'Nazario': df_nazario,
    'Nigerian_Fraud': df_nigerian,
    'SpamAssassin': df_spamAssassin,
}

print("=" * 60)
for name, df in datasets.items():
    print(f"\n{'='*20} {name} {'='*20}")
    print(f"  Shape     : {df.shape[0]:,} rows × {df.shape[1]} columns")
    print(f"  Columns   : {list(df.columns)}")
    print(f"  Label val : {df['label'].unique()}")
print("=" * 60)
print(f"\nTotal rows across all datasets: {sum(df.shape[0] for df in datasets.values()):,}")
```

### 1.2 — Missing Values Analysis

Null values are dangerous in text ML:
- A null **`body`** means we have NO content to analyze — that row must be **dropped**
- A null **`subject`** is totally normal (many legitimate emails have no subject) — fill with `""`
- Null `sender`, `receiver`, `date` are acceptable — we won't use those as final features

The heatmap below makes null patterns visually obvious. Yellow = null, dark = present.

```python
# 1.2 — Null value counts and heatmaps
fig, axes = plt.subplots(1, 5, figsize=(22, 5))
fig.suptitle("Missing Values per Dataset (yellow = null)", fontsize=14, fontweight='bold')

for ax, (name, df) in zip(axes, datasets.items()):
    # msno-style heatmap using seaborn
    null_matrix = df.isnull().astype(int)
    sns.heatmap(null_matrix, ax=ax, cbar=False, yticklabels=False,
                cmap='YlOrRd', vmin=0, vmax=1)
    ax.set_title(f"{name}\n({df.shape[0]:,} rows)", fontsize=10)
    ax.set_xticklabels(ax.get_xticklabels(), rotation=45, ha='right', fontsize=8)

plt.tight_layout()
plt.show()

# Summary table
print("\n=== NULL COUNTS PER DATASET ===")
for name, df in datasets.items():
    nulls = df.isnull().sum()
    if nulls.sum() > 0:
        print(f"\n{name}:")
        print(nulls[nulls > 0].to_string())
    else:
        print(f"\n{name}: No nulls")
```

## Phase 2: Data Cleaning & Schema Unification

Our 5 datasets have **inconsistent schemas** — some have `sender/receiver/date/urls`, others only have `subject/body/label`. We need to create one unified DataFrame before any ML work.

### Strategy
1. **Extract `has_url`** from the `urls` column (for datasets that have it) as a binary feature (0/1), then drop the column
2. **Drop metadata columns** (`sender`, `receiver`, `date`) — they're not available in all datasets and carry privacy risk
3. **Keep only**: `subject`, `body`, `label`, `has_url`
4. **Concatenate** all 5 DataFrames
5. **Handle nulls**: fill null subjects with `""`, drop rows with null body (no content = no signal)
6. **Drop duplicates** — some phishing emails appear in multiple datasets

### Why drop sender/date even though they have signal?
In a real phishing detector, you might have sender info. But for a **generalizable model** trained on mixed datasets, sender addresses are dataset-specific artifacts, not patterns. If Nazario phishing emails all come from certain domains, the model learns those specific domains instead of the email *content* — making it brittle on new data.

```python
# 2.1 — Extract has_url flag and normalize schemas

def extract_has_url(df):
    """Convert the urls column to a binary has_url flag."""
    if 'urls' in df.columns:
        # urls column contains integers (0 or 1) indicating URL presence
        return df['urls'].apply(lambda x: 1 if pd.notna(x) and str(x).strip() not in ('0', '', 'nan') else 0)
    else:
        return pd.Series(0, index=df.index)  # conservative: assume no URL info

# Build normalized copies of each dataset
dfs_normalized = []
for name, df in datasets.items():
    df_norm = pd.DataFrame({
        'subject': df['subject'],
        'body': df['body'],
        'label': df['label'],
        'has_url': extract_has_url(df),
        'source': name  # keep track of origin for debugging
    })
    dfs_normalized.append(df_norm)
    print(f"{name}: {len(df_norm):,} rows → has_url distribution: {df_norm['has_url'].value_counts().to_dict()}")

# 2.2 — Concatenate
df_all = pd.concat(dfs_normalized, ignore_index=True)
print(f"\nCombined shape: {df_all.shape}")
print(f"Label distribution:\n{df_all['label'].value_counts().to_string()}")
```

```python
# 2.3 — Handle nulls and duplicates

print(f"Before cleaning: {len(df_all):,} rows")
print(f"  Null subjects : {df_all['subject'].isnull().sum():,}")
print(f"  Null bodies   : {df_all['body'].isnull().sum():,}")

# Fill null subjects (normal — many emails have no subject)
df_all['subject'] = df_all['subject'].fillna('')

# Drop rows with null body (no text = no signal)
df_all = df_all.dropna(subset=['body'])
print(f"\nAfter dropping null bodies: {len(df_all):,} rows")

# Drop exact duplicates (same subject + body = same email)
df_all = df_all.drop_duplicates(subset=['subject', 'body'])
print(f"After deduplication: {len(df_all):,} rows")

# Reset index
df_all = df_all.reset_index(drop=True)

print(f"\nFinal label distribution:")
counts = df_all['label'].value_counts().sort_index()
for lbl, cnt in counts.items():
    name = 'Legitimate' if lbl == 0 else 'Phishing'
    pct = cnt / len(df_all) * 100
    print(f"  {name} (label={lbl}): {cnt:,} ({pct:.1f}%)")

print(f"\nFinal dataset shape: {df_all.shape}")
df_all.head(3)
```

## Phase 3: Text Preprocessing

Raw email text is extremely noisy. Before we can extract meaningful features or build vocabulary vectors, we need to **standardize the text**. Here's what we clean and why:

| Step | What | Why |
|------|------|-----|
| Lowercase | `"VERIFY YOUR ACCOUNT"` → `"verify your account"` | Same word shouldn't count as two different tokens |
| Remove HTML tags | `<a href="...">Click</a>` → `"Click"` | Email bodies often contain HTML; tags add noise |
| Extract URL flag | Count URLs before removing them | URLs are a strong phishing signal — we capture this as a feature |
| Replace URLs | Replace all URLs with the token `"URL"` | Keeps the *presence* of URLs as a signal without overfitting to specific domains |
| Remove special characters | Keep only letters, digits, spaces | Punctuation like `!!!` creates vocabulary noise |
| Collapse whitespace | `"  hello   world  "` → `"hello world"` | Inconsistent spacing from HTML stripping |

### Why NOT remove stopwords?
In general NLP, stopwords ("the", "is", "you") are removed. But for phishing, words like **"your"**, **"you"**, **"immediately"** carry strong signal. We'll keep them.

### Combining subject + body
The subject line is disproportionately important in phishing — it's crafted to trick the user into opening the email. By prepending the subject to the body, we give it more weight in the vocabulary.

```python
import re

def preprocess_text(text):
    """
    Clean a single email text string.
    Steps: lowercase → remove HTML → replace URLs → remove special chars → collapse whitespace
    """
    if not isinstance(text, str):
        return ""
    # 1. Lowercase
    text = text.lower()
    # 2. Remove HTML tags
    text = re.sub(r'<[^>]+>', ' ', text)
    # 3. Replace URLs with a token (presence is captured separately as has_url)
    text = re.sub(r'https?://\S+|www\.\S+', ' url ', text)
    # 4. Remove non-alphanumeric characters (keep spaces and letters/digits)
    text = re.sub(r'[^a-z0-9\s]', ' ', text)
    # 5. Collapse multiple whitespace
    text = re.sub(r'\s+', ' ', text).strip()
    return text

# Apply preprocessing
print("Preprocessing text... (may take ~1-2 minutes for large datasets)")
df_all['subject_clean'] = df_all['subject'].apply(preprocess_text)
df_all['body_clean'] = df_all['body'].apply(preprocess_text)

# Combine: subject prepended to body (subject gets proportionally more weight)
# We repeat subject twice to amplify its influence — a deliberate design choice
df_all['email_text'] = df_all['subject_clean'] + ' ' + df_all['subject_clean'] + ' ' + df_all['body_clean']

print("Done! Sample cleaned emails:")
print("\n--- PHISHING EXAMPLE ---")
phishing_sample = df_all[df_all['label'] == 1].iloc[0]
print(f"Subject (clean): {phishing_sample['subject_clean'][:100]}")
print(f"Body (clean):    {phishing_sample['body_clean'][:200]}")

print("\n--- LEGITIMATE EXAMPLE ---")
legit_sample = df_all[df_all['label'] == 0].iloc[0]
print(f"Subject (clean): {legit_sample['subject_clean'][:100]}")
print(f"Body (clean):    {legit_sample['body_clean'][:200]}")
```

## Phase 4: Feature Engineering

This is where domain knowledge about phishing becomes **competitive advantage**. Instead of just feeding raw text to a model (which only works well for deep learning), we engineer explicit numeric features that capture known phishing patterns.

We build **two types of features** that we'll combine:

### Type A: Handcrafted features (numeric, fast, interpretable)
Based on what we know phishing emails look like:
- They use **urgency language** ("verify", "account suspended", "act now")
- They're **full of capital letters** and exclamation marks for emphasis
- They reference **account/financial information** with lots of digits
- Short or repetitive text (Nigerian fraud style)

### Type B: TF-IDF text features (statistical, vocabulary-based)
**TF-IDF** (Term Frequency–Inverse Document Frequency) converts text into a numeric vector. For each word:
- **TF** = how often the word appears in THIS email
- **IDF** = how rare the word is across ALL emails

Result: common words (like "the") get low weight; rare, distinctive words (like "usaa", "password-expired") get high weight. We use **bigrams** (two-word phrases) too, so "verify account" is captured as a unit, not just "verify" and "account" separately.

### Combining both types
We use `scipy.sparse.hstack` to horizontally stack:
- The TF-IDF sparse matrix (5000 columns, most are zero for any given email)
- The dense numeric features (a few columns, all non-zero)

This gives each model access to both vocabulary patterns AND structural signals.

```python
# 4.1 — Handcrafted phishing-signal features (computed on RAW text, before cleaning)

# Urgency keywords — words deliberately used to create panic and prompt action
URGENCY_KEYWORDS = [
    'verify', 'update', 'confirm', 'urgent', 'immediately', 'suspend',
    'account', 'password', 'click', 'login', 'security', 'limited',
    'act now', 'expire', 'validate', 'reactivate', 'locked', 'blocked',
    'alert', 'attention', 'warning', 'important', 'unusual activity'
]

def compute_features(row):
    raw_text = str(row['subject']) + ' ' + str(row['body'])
    raw_lower = raw_text.lower()
    
    # Subject length
    subj_len = len(str(row['subject']))
    
    # Body length
    body_len = len(str(row['body']))
    
    # Word count (from cleaned text)
    word_count = len(str(row['email_text']).split())
    
    # Urgency score: count how many urgency keywords appear
    urgency_score = sum(1 for kw in URGENCY_KEYWORDS if kw in raw_lower)
    
    # Caps ratio: proportion of uppercase letters (phishing loves SHOUTING)
    letters = [c for c in raw_text if c.isalpha()]
    caps_ratio = sum(1 for c in letters if c.isupper()) / (len(letters) + 1)
    
    # Exclamation count
    exclamation_count = raw_text.count('!')
    
    # Digit ratio (fraud emails reference lots of account numbers/amounts)
    digit_ratio = sum(1 for c in raw_text if c.isdigit()) / (len(raw_text) + 1)
    
    # Unique word ratio: low uniqueness = repetitive/template text
    words = raw_lower.split()
    unique_word_ratio = len(set(words)) / (len(words) + 1)
    
    return pd.Series({
        'subject_len': subj_len,
        'body_len': body_len,
        'word_count': word_count,
        'urgency_score': urgency_score,
        'caps_ratio': caps_ratio,
        'exclamation_count': exclamation_count,
        'digit_ratio': digit_ratio,
        'unique_word_ratio': unique_word_ratio,
    })

print("Computing handcrafted features... (may take a moment)")
features_df = df_all.apply(compute_features, axis=1)
df_all = pd.concat([df_all, features_df], axis=1)

print("Done! Feature statistics by label:")
feature_cols = list(features_df.columns) + ['has_url']
print(df_all.groupby('label')[feature_cols].mean().round(3).to_string())
```

```python
# 4.2 — Visualize feature distributions: Do our engineered features separate the classes?
# This is the validation step — if a feature doesn't separate classes, it adds noise.

fig, axes = plt.subplots(2, 4, figsize=(20, 9))
axes = axes.flatten()

plot_features = ['urgency_score', 'caps_ratio', 'exclamation_count',
                 'digit_ratio', 'unique_word_ratio', 'has_url',
                 'subject_len', 'body_len']

for ax, feat in zip(axes, plot_features):
    for label, color, name in [(0, '#2ecc71', 'Legitimate'), (1, '#e74c3c', 'Phishing')]:
        subset = df_all[df_all['label'] == label][feat]
        # Cap at 99th percentile for readability
        cap = subset.quantile(0.99)
        subset_capped = subset[subset <= cap]
        ax.hist(subset_capped, bins=40, alpha=0.6, color=color, label=name, density=True)
    ax.set_title(feat, fontweight='bold')
    ax.legend(fontsize=8)
    ax.set_ylabel("Density")

plt.suptitle("Feature Distributions: Phishing vs. Legitimate\n(Separation = useful feature)", 
             fontsize=13, fontweight='bold')
plt.tight_layout()
plt.show()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
import scipy.sparse as sp
from sklearn.preprocessing import StandardScaler

# 4.3 — TF-IDF vectorization
# max_features=5000: keep only the 5000 most informative words/bigrams
# ngram_range=(1,2): use both unigrams ("verify") and bigrams ("verify account")
# min_df=2: ignore words that appear in fewer than 2 documents (likely typos/noise)
# sublinear_tf=True: use log(1+tf) instead of raw tf — dampens effect of very frequent words

print("Fitting TF-IDF vectorizer...")
tfidf = TfidfVectorizer(
    max_features=5000,
    ngram_range=(1, 2),
    min_df=2,
    sublinear_tf=True
)
X_tfidf = tfidf.fit_transform(df_all['email_text'])
print(f"TF-IDF matrix shape: {X_tfidf.shape}")
print(f"  → {X_tfidf.shape[0]:,} emails × {X_tfidf.shape[1]:,} word/bigram features")
print(f"  → Sparsity: {(1 - X_tfidf.nnz / (X_tfidf.shape[0] * X_tfidf.shape[1])):.1%} of values are zero")

# 4.4 — Combine TF-IDF with handcrafted features
HANDCRAFTED_COLS = ['subject_len', 'body_len', 'word_count', 'urgency_score',
                     'caps_ratio', 'exclamation_count', 'digit_ratio', 
                     'unique_word_ratio', 'has_url']

# Scale numeric features to [0,1] range so they don't dominate TF-IDF
scaler = StandardScaler()
X_numeric = scaler.fit_transform(df_all[HANDCRAFTED_COLS])
X_numeric_sparse = sp.csr_matrix(X_numeric)

# Stack horizontally: each row = one email, columns = TF-IDF + numeric features
X = sp.hstack([X_tfidf, X_numeric_sparse])
y = df_all['label'].values

print(f"\nFinal feature matrix shape: {X.shape}")
print(f"  → {X.shape[1]} total features ({X_tfidf.shape[1]} TF-IDF + {len(HANDCRAFTED_COLS)} handcrafted)")
print(f"\nTop 20 most informative TF-IDF features by IDF weight:")
feature_names = tfidf.get_feature_names_out()
idf_scores = tfidf.idf_
top_idx = idf_scores.argsort()[-20:][::-1]
for idx in top_idx:
    print(f"  '{feature_names[idx]}': {idf_scores[idx]:.3f}")
```

## Phase 5: Class Imbalance & Train/Test Split

### Why imbalance is dangerous

Suppose we have 80,000 legitimate emails and 20,000 phishing emails (4:1 ratio). A model that **always predicts "legitimate"** achieves **80% accuracy** with zero actual usefulness. This is why we need special handling.

### Our approach: `class_weight='balanced'`
Most sklearn models accept a `class_weight='balanced'` parameter. This makes the model internally penalize mistakes on the minority class more heavily — equivalent to oversampling phishing emails during training, but without actually duplicating data.

**Formula**: weight for class `i` = `total_samples / (n_classes × count_of_class_i)`

### Train/Test split: why `stratify=y`?
Without stratification, random splitting could put all phishing emails in train and none in test (by chance, especially if phishing is rare). `stratify=y` guarantees both splits have the same class ratio.

```python
from sklearn.model_selection import train_test_split

# 5.1 — Check class imbalance ratio
class_counts = pd.Series(y).value_counts().sort_index()
n_legit = class_counts.get(0, 0)
n_phish = class_counts.get(1, 0)
imbalance_ratio = n_legit / n_phish if n_phish > 0 else float('inf')

print(f"Class distribution:")
print(f"  Legitimate (0): {n_legit:,} ({n_legit/len(y)*100:.1f}%)")
print(f"  Phishing   (1): {n_phish:,} ({n_phish/len(y)*100:.1f}%)")
print(f"  Imbalance ratio: {imbalance_ratio:.2f}:1")

if imbalance_ratio > 3:
    print(f"\n⚠️  Significant imbalance detected ({imbalance_ratio:.1f}:1).")
    print("   We'll use class_weight='balanced' in all models.")
else:
    print(f"\n✓ Imbalance is manageable ({imbalance_ratio:.1f}:1).")

# XGBoost's equivalent of class_weight='balanced'
xgb_scale_pos_weight = n_legit / n_phish
print(f"\nXGBoost scale_pos_weight = {xgb_scale_pos_weight:.2f}")

# 5.2 — Stratified train/test split (80% train, 20% test)
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.20,
    random_state=42,
    stratify=y  # ensures same class ratio in both splits
)

print(f"\nTrain set: {X_train.shape[0]:,} samples")
print(f"  Legitimate: {(y_train == 0).sum():,} | Phishing: {(y_train == 1).sum():,}")
print(f"\nTest set: {X_test.shape[0]:,} samples")
print(f"  Legitimate: {(y_test == 0).sum():,} | Phishing: {(y_test == 1).sum():,}")
```

## Phase 7: Model Training

We train 3 models, chosen deliberately to represent different algorithmic approaches:

### Model 1: Logistic Regression
**What it does**: Learns a linear decision boundary — it finds a weighted sum of features that best separates phishing from legitimate. Think of it as learning: "if urgency_score > X AND has the word 'verify' → phishing".

**Why start here**: It's the best baseline for text classification with TF-IDF. Fast, interpretable (you can look at the coefficients to understand what drives predictions), and often surprisingly competitive.

**Key parameters**:
- `C=1.0`: Regularization strength — higher C = less regularization = model can fit training data more closely (risk of overfitting)
- `class_weight='balanced'`: compensates for imbalance
- `max_iter=1000`: LR uses iterative optimization; needs enough iterations to converge

---

### Model 2: Random Forest
**What it does**: Trains 200 decision trees, each on a random subset of data and features. The final prediction is a majority vote across all trees. This "ensemble" approach is much more robust than a single tree.

**Why include it**: Random Forest naturally handles **non-linear relationships** (unlike LR). It also provides **feature importance** scores, which will help us understand which signals matter most.

**Key parameters**:
- `n_estimators=200`: 200 trees. More = better but slower
- `max_depth=20`: Prevents trees from memorizing training data
- `class_weight='balanced'`

---

### Model 3: Gradient Boosting (XGBoost)

## Phase 8: Evaluation & Comparison

### Understanding the metrics

For phishing detection, metrics have very concrete meaning:

| Metric | Formula | Meaning in our context |
|--------|---------|----------------------|
| **Recall** | TP / (TP + FN) | Of all actual phishing emails, what % did we catch? **Missed phishing = user gets harmed** |
| **Precision** | TP / (TP + FP) | Of all emails we flagged as phishing, what % really were? Too low = users ignore alerts (cry wolf) |
| **F1-score** | 2 × (P×R)/(P+R) | Harmonic mean — the right single metric when recall and precision both matter |
| **ROC-AUC** | Area under ROC curve | Model's overall discrimination ability (threshold-independent) |
| **Accuracy** | (TP+TN)/(all) | Least useful here — misleading with imbalanced data |

### The recall vs. precision tradeoff
In security, **recall is usually more important** — better to flag a few extra suspicious emails (false alarms are annoying) than to miss actual phishing (user gets hacked). However, if false alarm rate is too high, users start ignoring ALL alerts, making the system useless. We need to find the right balance.

```python
from sklearn.metrics import (
    classification_report, confusion_matrix, roc_auc_score,
    roc_curve, precision_recall_curve, average_precision_score, accuracy_score
)

models = {
    'Logistic Regression': lr,
    'Random Forest': rf,
    'Gradient Boosting': gb,
}

# Collect predictions and probabilities for all models
results = {}
for name, model in models.items():
    y_pred = model.predict(X_test)
    y_proba = model.predict_proba(X_test)[:, 1]  # probability of being phishing
    results[name] = {'y_pred': y_pred, 'y_proba': y_proba}

# 8.1 — Classification reports (precision, recall, F1 per class)
print("=" * 65)
for name, res in results.items():
    print(f"\n{'='*65}")
    print(f"  {name}")
    print(f"{'='*65}")
    print(classification_report(y_test, res['y_pred'],
                                target_names=['Legitimate', 'Phishing']))
```

```python
# 8.2 — Confusion matrices: visualize what kinds of errors each model makes

fig, axes = plt.subplots(1, 3, figsize=(18, 5))
class_labels = ['Legitimate', 'Phishing']

for ax, (name, res) in zip(axes, results.items()):
    cm = confusion_matrix(y_test, res['y_pred'])
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', ax=ax,
                xticklabels=class_labels, yticklabels=class_labels,
                annot_kws={'size': 14})
    ax.set_title(f"{name}", fontweight='bold')
    ax.set_ylabel("Actual")
    ax.set_xlabel("Predicted")
    
    # Add interpretation below each matrix
    tn, fp, fn, tp = cm.ravel()
    missed = fn / (tp + fn) * 100 if (tp + fn) > 0 else 0
    false_alarm = fp / (tn + fp) * 100 if (tn + fp) > 0 else 0
    ax.set_xlabel(
        f"Predicted\n[Missed phishing: {missed:.1f}% | False alarms: {false_alarm:.1f}%]"
    )

plt.suptitle("Confusion Matrices\n(Missed phishing = dangerous | False alarms = annoying)", 
             fontsize=12, fontweight='bold')
plt.tight_layout()
plt.show()
```

```python
# 8.3 — ROC Curves + Precision-Recall Curves (all 3 models overlaid)
# ROC curve: plots True Positive Rate vs False Positive Rate at every threshold
# PR curve: plots Precision vs Recall at every threshold (better for imbalanced data)

fig, axes = plt.subplots(1, 2, figsize=(16, 6))
colors_models = ['#3498db', '#e67e22', '#9b59b6']

for (name, res), color in zip(results.items(), colors_models):
    y_proba = res['y_proba']
    
    # ROC curve
    fpr, tpr, _ = roc_curve(y_test, y_proba)
    auc = roc_auc_score(y_test, y_proba)
    axes[0].plot(fpr, tpr, color=color, lw=2, label=f"{name} (AUC={auc:.3f})")
    
    # Precision-Recall curve
    precision, recall, _ = precision_recall_curve(y_test, y_proba)
    ap = average_precision_score(y_test, y_proba)
    axes[1].plot(recall, precision, color=color, lw=2, label=f"{name} (AP={ap:.3f})")

# ROC baseline
axes[0].plot([0, 1], [0, 1], 'k--', lw=1, label='Random baseline')
axes[0].set_xlabel("False Positive Rate (False Alarm Rate)")
axes[0].set_ylabel("True Positive Rate (Recall)")
axes[0].set_title("ROC Curves", fontweight='bold')
axes[0].legend()
axes[0].grid(alpha=0.3)

# PR baseline
baseline_pr = (y_test == 1).mean()
axes[1].axhline(y=baseline_pr, color='k', linestyle='--', lw=1, label=f'Random baseline ({baseline_pr:.2f})')
axes[1].set_xlabel("Recall (% of phishing caught)")
axes[1].set_ylabel("Precision (% of flagged that are real phishing)")
axes[1].set_title("Precision-Recall Curves\n(More meaningful for imbalanced data)", fontweight='bold')
axes[1].legend()
axes[1].grid(alpha=0.3)

plt.tight_layout()
plt.show()
```

```python
# 8.4 — Summary comparison table: all models side by side

from sklearn.metrics import f1_score, precision_score, recall_score

summary_rows = []
for name, res in results.items():
    y_pred = res['y_pred']
    y_proba = res['y_proba']
    summary_rows.append({
        'Model': name,
        'Accuracy': accuracy_score(y_test, y_pred),
        'Precision (phishing)': precision_score(y_test, y_pred),
        'Recall (phishing)': recall_score(y_test, y_pred),
        'F1 (phishing)': f1_score(y_test, y_pred),
        'ROC-AUC': roc_auc_score(y_test, y_proba),
        'Avg Precision (PR)': average_precision_score(y_test, y_proba),
    })

df_summary = pd.DataFrame(summary_rows).set_index('Model')

print("\n" + "="*75)
print("MODEL COMPARISON SUMMARY")
print("="*75)
print(df_summary.round(4).to_string())
print("="*75)

# Highlight winner per metric
print("\nBest model per metric:")
for col in df_summary.columns:
    best = df_summary[col].idxmax()
    val = df_summary[col].max()
    print(f"  {col:30s}: {best} ({val:.4f})")
```

### How to choose the winner
- If **recall** is your priority (catch every phishing email, accept some false alarms): pick the model with highest recall on phishing class
- If **balance** matters (both catching phishing AND not annoying users): use F1-score as the tiebreaker
- For **deployment**, also consider: inference speed (LR is fastest), explainability (LR coefficients are readable), and memory (LR << RF << XGB)

### Potential improvements to explore
- **Threshold tuning**: models predict probabilities — you can lower the decision threshold from 0.5 to 0.3 to trade precision for recall
- **SMOTE**: synthetic oversampling of the minority class (alternative to class_weight)
- **Cross-validation**: use `StratifiedKFold` to get more robust metric estimates
- **More features**: sender domain spoofing patterns, HTML link count, reply-to mismatch
- **Fine-tuned BERT**: for the highest possible accuracy, a small distilBERT model trained on email text would outperform all of the above
